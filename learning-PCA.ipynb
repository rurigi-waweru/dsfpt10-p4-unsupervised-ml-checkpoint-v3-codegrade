```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. Load the data
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# 2. Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Apply PCA (reduce to 2 components for visualization)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 4. Plot the results
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], alpha=0.7, color=colors[i], label=target_name)
    
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Iris Dataset')
plt.legend()
plt.grid(True)
plt.show()
```

üß† What This Does:
Standardizes the 4 original features (sepal length, sepal width, petal length, petal width).

Reduces the data from 4D to 2D using PCA.

Plots the new 2D representation, coloring by species.

**üßê Interpretation:**
Even though we reduced the data from 4 features to 2, we still see clear clustering patterns between the different iris species ‚Äî which shows PCA preserved much of the structure.
